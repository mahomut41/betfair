```python
import os
import shutil

# List to store destination file paths
destination_file_paths = []

# Function to find the largest .bz2 file in a directory
def find_largest_bz2_file(directory):
    largest_files = []
    largest_size = 0

    # Iterate over all files in the directory
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".bz2"):
                file_path = os.path.join(root, file)
                file_size = os.path.getsize(file_path)
                if file_size > largest_size:
                    largest_size = file_size
                    largest_files = [file_path]
                elif file_size == largest_size:
                    largest_files.append(file_path)

    # If there are multiple files with the same largest size, choose the one with the same name as the last folder
    if len(largest_files) > 1:
        folder_name = os.path.basename(directory)
        for file_path in largest_files:
            if os.path.basename(file_path) == folder_name + ".bz2":
                return file_path

    # If there is only one largest file or none match the folder name, return the first one
    return largest_files[0] if largest_files else None

# Function to copy the largest .bz2 file to the destination directory
def copy_largest_bz2_file(source_file, destination_dir):
    if source_file:
        filename = os.path.basename(source_file)
        destination_file = os.path.join(destination_dir, filename)
        shutil.copy2(source_file, destination_file)
        print(f"Copied '{source_file}' to '{destination_file}'")
        destination_file_paths.append(destination_file)  # Store destination file path
        return 1
    return 0

# Recursively find the largest .bz2 file in each child folder and copy it to the destination directory
def find_and_copy_largest_bz2_files(directory, destination_dir):
    total_files_copied = 0
    for root, dirs, files in os.walk(directory):
        if not dirs:  # If the current directory has no subdirectories, it's a leaf node
            largest_bz2_file = find_largest_bz2_file(root)
            total_files_copied += copy_largest_bz2_file(largest_bz2_file, destination_dir)
    return total_files_copied

# Main function to start the process
def main():
    source_dir = "/Users/noahroni/Documents/Data-2/BASIC/2020/May"
    destination_dir = "/Users/noahroni/Documents/event_files"

    if not os.path.exists(destination_dir):
        os.makedirs(destination_dir)

    total_files_copied = find_and_copy_largest_bz2_files(source_dir, destination_dir)
    print(f"Total files copied: {total_files_copied}")

    # Print destination file paths
    print("Destination file paths:")
    #for file_path in destination_file_paths:
        #print(file_path)

# Run the main function
if __name__ == "__main__":
    main()

#print(destination_file_paths[0])
```
