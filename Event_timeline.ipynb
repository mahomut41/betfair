```python
import matplotlib.pyplot as plt
import os

# Sample data for demonstration
teams = ['PSS Sleman', 'Arema Cronus']
events = [15, 29, 39, 41]  # Match times for goals
goals = ['Goal', 'Goal', 'Goal']  # Types of events (e.g., Goal, Penalty, etc.)

# Plotting the timeline
plt.figure(figsize=(12, 4))

# Plotting goals for each team
for team in teams:
    team_events = [events[i] for i in range(len(events)) if teams[i % len(teams)] == team]
    team_goals = [goals[i] for i in range(len(goals)) if teams[i % len(teams)] == team]
    plt.scatter(team_events, [0] * len(team_events), marker='o', label=team, color='blue' if team == 'PSS Sleman' else 'orange')  # Plotting on the line at y=0
    for i, txt in enumerate(team_goals):
        plt.annotate(txt, (team_events[i], 0), xytext=(0, 10), textcoords='offset points', ha='center', fontsize=8)  # Adjusting annotation position

# Adding a line in the middle
plt.hlines(0, min(events), max(events), colors='red', linestyles='dashed')

# Adding labels and title
plt.xlabel('Match Time (minutes)')
plt.ylabel('Events')
plt.title('Timeline of Events')

# Adding legend
plt.legend()

# Displaying the timeline plot
plt.grid(True)
plt.yticks([])
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data for demonstration
teams = ['PSS Sleman', 'Arema Cronus']
yellow_cards = [3, 2]  # Number of yellow cards for each team
corner_kicks = [5, 3]  # Number of corner kicks for each team

# Plotting the data
plt.figure(figsize=(10, 6))

# Plotting corner kicks
plt.barh([team + ' (Corner Kicks)' for team in teams], corner_kicks, color='blue', label='Corner Kicks')

# Plotting yellow cards
plt.barh([team + ' (Yellow Cards)' for team in teams], yellow_cards, color='orange', label='Yellow Cards')

# Adding yellow card numbers on the bars
for i, value in enumerate(yellow_cards):
    plt.text(value + 0.1, i, str(value), va='center')

# Adding corner kick numbers on the bars
for i, value in enumerate(corner_kicks):
    plt.text(value + 0.1, i + len(teams), str(value), va='center')

# Adding labels and title
plt.xlabel('Count')
plt.title('Yellow Cards and Corner Kicks per Team')

# Adding legend
plt.legend()

# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data for demonstration
teams = ['Team A']
values_left = [5]  # Values for Team A
values_right = [3]  # Values for Team B

# Convert values to percentages
total = sum(values_left + values_right)
percentages_left = [val / total * 100 for val in values_left]
percentages_right = [val / total * 100 for val in values_right]

# Plotting the data
plt.figure(figsize=(8, 0.1))

# Creating the left stack (representing Team A)
plt.barh(teams, percentages_left, color='blue', label='Team A')

# Creating the right stack (representing Team B)
plt.barh(teams, percentages_right, left=percentages_left, color='red', label='Team B')

# Adding the name of Team A at the right side
plt.annotate('Team A', xy=(1, 0), xytext=(-10, 5), ha='right', va='top', textcoords='offset points')

# Adding the name of Team A at the right side
plt.annotate('Team B', xy=(1, 0), xytext=(485, 5), ha='right', va='top', textcoords='offset points')

# Adding labels and title
plt.xlabel('Percentage')
plt.title('Corner Kicks')

# Set the x-axis limits to start from 0 and end at 100
plt.xlim(0, 100)

# Remove y-axis ticks
plt.yticks([])

# Displaying the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import os

# Extracted data from the HTML code
events = [
    {"type": "KickOff", "position": 0},
    {"type": "Goal", "position": 2.22, "team": "home"},
    {"type": "Goal", "position": 22.22, "team": "home"},
    {"type": "Goal", "position": 71.11, "team": "away"},
    {"type": "SecondHalfKickOff", "position": 100},
    {"type": "FirstHalfEnd", "position": 2.22},
    {"type": "Goal", "position": 15.56, "team": "home"},
    {"type": "Goal", "position": 20.00, "team": "away"},
    {"type": "Goal", "position": 62.22, "team": "home"},
    {"type": "YellowCard", "position": 77.78, "team": "home"},
    {"type": "Goal", "position": 80.00, "team": "away"},
    {"type": "YellowCard", "position": 100, "team": "away"},
    {"type": "SecondHalfEnd", "position": 100},
]

# Define event types and their corresponding colors
event_colors = {
    "KickOff": "blue",
    "Goal": "green",
    "YellowCard": "yellow",
    "RedCard": "red",
    "FirstHalfEnd": "purple",
    "SecondHalfKickOff": "orange",
    "SecondHalfEnd": "brown",
}

# Plot the timeline
plt.figure(figsize=(10, 5))
plt.plot([0, 100], [0, 0], color='black')  # Draw the timeline

for event in events:
    color = event_colors.get(event["type"], "black")
    if event["type"] == "Goal":
        marker = "o"
    elif event["type"] == "YellowCard":
        marker = "^"
    else:
        marker = "s"
    plt.scatter(event["position"], 0, color=color, marker=marker, s=100)
    plt.text(event["position"], 0.05, event["type"], ha='center')

# Add a gap at 45 min mark  
plt.axvspan(45, 45.1, color='r', zorder=-1)

plt.yticks([])
plt.xlabel("Timeline")
plt.title("Soccer Match Timeline")
plt.grid(True)
plt.xlim(0, 100)
plt.ylim(-1, 1)

# Save the graph in a folder
output_folder = 'graphs'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

output_file = os.path.join(output_folder, 'Soccer Match Timeline_graph.png')
plt.savefig(output_file)
plt.show()
```
